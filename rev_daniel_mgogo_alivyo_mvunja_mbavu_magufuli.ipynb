```python
import whisper
from pytube import YouTube
```

```python
model = whisper.load_model('large-v2')
```

```python
youtube_video_url = "https://www.youtube.com/watch?v=EHfwCz35iAg&t=407s"
youtube_video = YouTube(youtube_video_url)
```

```python
youtube_video.title
```

```python
dir(youtube_video)
```

```python
youtube_video.streams
```

```python
for stream in youtube_video.streams:
  print(stream)
```

```python
streams = youtube_video.streams.filter(only_audio=True)
streams
```

```python
stream = streams.last()
stream
```

```python
stream.download(filename='rev_daniel_mgogo_alivyo_mvunja_mbavu_magufuli.mp4')
```

```python
# do the transcription
import datetime

# save a timestamp before transcription
t1 = datetime.datetime.now()
print(f"started at {t1}")

# do the transcription
output = model.transcribe("rev_daniel_mgogo_alivyo_mvunja_mbavu_magufuli.mp4")

# show time elapsed after transcription is complete.
t2 = datetime.datetime.now()
print(f"ended at {t2}")
print(f"time elapsed: {t2 - t1}")
```

```python
from googletrans import Translator
```

```python
# The text to translate
text = ["rev_daniel_mgogo_alivyo_mvunja_mbavu_magufuli.txt"]
```

```python
print(text)
length = len(text.split())
print(length)
```

```python
# Create a translator object
translator = Translator()
```

```python
# Translate the text
rev_daniel_mgogo_alivyo_mvunja_mbavu_magufuli_translated = translator.translate(text, src='sw', dest='en')
```

```python
# Print the translated text
print(rev_daniel_mgogo_alivyo_mvunja_mbavu_magufuli_translated.text)
```
